```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from qutip import *
```

```python
res = qload('rp_6')
```

```python
fres = ptrace(res.states[-1], [0,1,2,3])
```

```python
a = [0]*16
for i in range(2):
    for j in range(2):
        for k in range(2):
            for l in range(2):
                a[8*i+4*j+2*k+l] = expect(fres, tensor(basis(2,i),basis(2,j),basis(2,k),basis(2,l)))
```

```python
plt.plot(a)
```
