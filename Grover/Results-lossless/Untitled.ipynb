```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from qutip import *
```

```python
res = []

for i in range(7): #7
    res.append(qload('it_{}'.format(i+1)))
```

```python
np.abs(res[3].states[-1][15][0,0])**2
res[3].states[-1][15]
```

```python
a = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]

for i in range(2):
    for j in range(2):
        for k in range(2):
            for l in range(2):
                for m in range(7): #7
                    a[8*i+4*j+2*k+l].append(np.abs(res[m].states[-1][8*i+4*j+2*k+l][0,0]))
```

```python
sum = 0
for i in range(16):
    sum += (a[i][-1])**2
    
print(sum)
```

```python
fig, axes = plt.subplots(1, 1, sharex=True, figsize=(12,8))

for i in range(16):
    if i == 15:
        axes.plot(np.abs([1/np.sqrt(16)]+a[i])**2, c='b', label='Estado objetivo')
    elif i == 0:
        axes.plot(np.abs([1/np.sqrt(16)]+a[i])**2, c='k', label='Otros estados')
    else:
        axes.plot(np.abs([1/np.sqrt(16)]+a[i])**2, c='k')
axes.set_xlabel('Iteración')
axes.set_ylabel('Probabilidad')
axes.set_ylim(0,1)
axes.legend()
```
