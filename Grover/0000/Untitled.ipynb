```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from qutip import *
```

```python
res = []

for i in range(7): #7
    res.append(qload('it_{}'.format(i+1)))
```

```python
res[3].states[-1][0,0]
```

```python
a = [[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16],[1/16]]

for i in range(2):
    for j in range(2):
        for k in range(2):
            for l in range(2):
                for m in range(7): #7
                    a[8*i+4*j+2*k+l].append(np.abs(res[m].states[-1][8*i+4*j+2*k+l,8*i+4*j+2*k+l]))
```

```python
[a[i][-1] for i in range(16)]
```

```python
fig, axes = plt.subplots(1, 1, sharex=True, figsize=(12,8))

for i in range(16):
    if i == 0:
        axes.plot(np.abs(a[i]), c='b', label='ω = 0 (Estado base)')
    elif i == 1:
        axes.plot(np.abs(a[i]), c='k', label='Otros estados')
    else:
        axes.plot(np.abs(a[i]), c='k')
axes.set_xlabel('Iteración')
axes.set_ylabel('Probabilidad')
axes.set_ylim(0,1)
axes.legend()
fig.savefig('../../img/grover0000loss.eps', format='eps')
```
