```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from qutip import *
```

```python
res = []

for i in range(7):
    res.append(qload('it_{}'.format(i+1)))
```

```python
a = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]

for i in range(2):
    for j in range(2):
        for k in range(2):
            for l in range(2):
                for m in range(7):
                    a[8*i+4*j+2*k+l].append(expect(res[m].states[-1], tensor(basis(2,i),basis(2,j),basis(2,k),basis(2,l))))
```

```python
for i in range(16):
    plt.plot(a[i])
```
